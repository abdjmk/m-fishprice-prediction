```python
columns = ['LANDED_WEIGHT_KG', 'LIVE_WEIGHT_KG', 'VALUE_MEXICAN_PESOS']

for column in columns:
    # Histogram of a numerical column
    sns.histplot(data[column], bins=30)  # Adjust the number of bins as per your preference
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Scatter plot of two numerical columns
plt.scatter(data['LIVE_WEIGHT_KG'], data['LANDED_WEIGHT_KG'])
plt.xlabel('LIVE_WEIGHT_KG')
plt.ylabel('LANDED_WEIGHT_KG')
plt.title('Scatter Plot')
plt.show()
```

```python
# Scatter plot of two numerical columns
plt.scatter(data['LANDED_WEIGHT_KG'], data['VALUE_MEXICAN_PESOS'])
plt.xlabel('LANDED_WEIGHT_KG')
plt.ylabel('VALUE_MEXICAN_PESOS')
plt.title('Scatter Plot')
plt.show()
```

```python
# Scatter plot of two numerical columns
plt.scatter(data['LIVE_WEIGHT_KG'], data['VALUE_MEXICAN_PESOS'])
plt.xlabel('LIVE_WEIGHT_KG')
plt.ylabel('VALUE_MEXICAN_PESOS')
plt.title('Scatter Plot')
plt.show()
```

```python
columns2 = ['YEAR', 'COAST', 'ENTITY', 'MONTH', 'GROUP', 'MAIN_SPECIES_NAME', 'ORIGIN',
            'ORIGIN_OF_THE_SHRIMP', 'WATER_TYPE', 'DESTINATION', 'AQUACULTURE']

# Set the color palette
sns.set_palette('cool')
for column in columns2:
    column_counts = data[column].value_counts()
    
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=column_counts.index, y=column_counts)
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    
    # Add labels to the bars
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', color='black', fontsize=8, fontweight='light')
    
    plt.show()
```

```python
# Box plot to visualize distribution and outliers
columns3 = ['YEAR', 'COAST', 'ENTITY', 'MONTH', 'GROUP', 'MAIN_SPECIES_NAME', 'ORIGIN',
            'ORIGIN_OF_THE_SHRIMP', 'WATER_TYPE', 'DESTINATION', 'AQUACULTURE',
            'LANDED_WEIGHT_KG', 'LIVE_WEIGHT_KG', 'VALUE_MEXICAN_PESOS']

y_column = 'VALUE_MEXICAN_PESOS'

for column in columns3:
    if column == y_column:
        continue
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=column, y=y_column, color='black')
    plt.title(f'{column} vs {y_column}')
    plt.xlabel(column)
    plt.ylabel(y_column)
    
    plt.show()
```
